#correlation with categorical

```python
from scipy.stats import chi2_contingency
```

```python
cat_features = ['term', 'grade', 'sub_grade', 'emp_title', 'emp_length',
       'home_ownership', 'issue_d', 'pymnt_plan', 'purpose', 'title',
       'zip_code', 'addr_state', 'earliest_cr_line', 'initial_list_status',
       'last_pymnt_d', 'last_credit_pull_d', 'application_type', 'issue_month',
       'last_pymnt_month', 'last_credit_pull_month',
       'earliest_cr_line_months','term_months']

error = 0.05

for column in cat_features:
    table = load_data[[column, 'loan_status']].pivot_table(index='loan_status',
                                                   columns=column,
                                                   aggfunc=lambda x: len(x))
    _, p_value, _, _ = chi2_contingency(table)

    print('Variable:', column)
    print('p-value: ', p_value)
    if p_value < error:
        print('Variabel ' + column + ' dan loan_status berkorelasi')
    else:
        print('Variabel ' + column + ' dan loan_status tidak berkorelasi')
    print('')
```
